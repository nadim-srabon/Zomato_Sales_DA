## deleteing redeundant columns from the dataset

```python
## deleteing redeundant columns from the dataset

columns_to_drop = ['url', 'address','votes',
       'phone', 'location', 'rest_type','cuisines', 'reviews_list', 'menu_item',
       'listed_in(type)', 'listed_in(city)']
columns_to_keep = ['name', 'online_order', 'book_table', 'rate','dish_liked','approx_cost(for two people)']
```

```python
len(columns_to_drop)
```

```python
len(columns_to_keep)
```

```python
modify_data = data[columns_to_keep]
modify_data

## or drop

data.drop(columns= columns_to_drop,inplace = True)
```

## rename the columns of the data 1 by 1

```python
data.rename(columns={'name':'name123','online_order':'online_order123'}) ## not committed unless inplace is true
```

```python
data.columns[0].capitalize()
```

```python
for i in data.columns:
    print(i.capitalize())
```

```python
new_columns_names = []
for i in data.columns:
    new_columns_names.append(i.capitalize())
```

```python
new_columns_names
```

```python
data.columns = new_columns_names
```

```python
data
```

## droping duplicates 

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

### here is some indexing problem which we need to reset 

```python
data.reset_index(drop=True, inplace=True)
```

```python
data
```

## cleaning data column by column

```python
data["Rate"]
```

```python
data['Rate'] = data["Rate"].str.replace(' ','')
```

```python
data['Rate'] =data["Rate"].str.replace('outof','/')
```

```python
data
```

## non-numeric column to numeric column

```python
data['Online_order'] = data['Online_order'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
data
```

## exporting the clean datasets

```python
data.to_csv('zomato_cleaned.csv', index=False)
```
